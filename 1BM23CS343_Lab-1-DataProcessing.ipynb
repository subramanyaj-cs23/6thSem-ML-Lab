```python
import pandas as pd

dataset = pd.read_csv('diabetes.csv')

print(dataset.head())

print(dataset.describe())

un = dataset['AGE'].unique()
print(un)
print(len(un))

missing_arr = dataset.isnull().sum()

missing_columns = missing_arr[missing_arr > 0]
print("Columns with Missing Values:\n", missing_columns)

```
